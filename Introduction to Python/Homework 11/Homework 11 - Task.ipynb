```python
import cmath
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
from sympy.plotting import plot
!matplotlib inline
```

```python
f = lambda x: -5*x**2 + 10*x - 150
f_x = lambda x: [f(i) for i in x]
```

```python
x = np.arange(-1000, 1000, 0.1)
plt.plot(x, f_x(x));
```

```python
x = np.arange(-5, 5, 0.1)
plt.plot(x, f_x(x));
```

Вычислим корни функции:

```python
def complex_solve(a, b, c):
    d = b**2 - 4*a*c
    x = [(-b-cmath.sqrt(d))/(2*a), (-b+cmath.sqrt(d))/(2*a)]
    return x

print(complex_solve(-5, 10, -150))
```

```python
x = Symbol('x')
fx = -5*x**2 + 10*x - 150
fx
```

```python
plot(fx);
```

Вычислим корни через solve

```python
solve(fx, x)
```

Функция f>0 на интервале:

```python
solve(fx>0)
```

Функция f<0 на интервале:

```python
solve(fx<0)
```

```python
df = diff(fx)
df
```

```python
plot(df);
```

Для нахождения вершины приравняем производную нулю:

```python
x_d = solve(df, x)
x_d
```

Значения агрумента и функции в вершине:

```python
x_d[0], f(x_d[0])
set(((x_d[i], f(x_d[i])) for i in range(len(x_d))))
```

Функция возрастает на интервале:

```python
solve(df>0)
```

Функция убывает на интервале:

```python
solve(df<0)
```
